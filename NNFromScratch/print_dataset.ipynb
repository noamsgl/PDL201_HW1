# Explore Dataset



```python
import numpy as np
import seaborn as sns
from scipy.io import loadmat
# np.set_printoptions(threshold=sys.maxsize)
```

```python
filenames = ['NNdata/SwissRollData.mat',
             'NNdata/PeaksData.mat',
             'NNdata/GMMData.mat']

# select dataset index:
# SwissRoll = 0
# Peaks = 1
# GMM = 2
index = 2

x = loadmat(filenames[index])
Xt, Xv, Ct, Cv, = x['Yt'], x['Yv'], x['Ct'], x['Cv']
```

```python
print("Xt", Xt.shape)
print(Xt)
```

```python
print("Ct", Ct.shape)
print(Ct)
```

```python
Y = np.argmax(Ct, axis=0)
print("Y=Labels", np.shape(Y))
print(Y)

```

```python
(unique, counts) = np.unique(Y, return_counts=True)
histogram = dict(zip(unique, counts))
print("Histogram")
print(histogram)

```

```python
sns.scatterplot(x=Xt[0], y=Xt[1], hue=Y)
```
