# Explore Dataset



```python
import numpy as np
import seaborn as sns
from scipy.io import loadmat
# np.set_printoptions(threshold=sys.maxsize)
```

```python
filenames = ['../NNdata/SwissRollData.mat',
             '../NNdata/PeaksData.mat',
             '../NNdata/GMMData.mat'
             ]

# select dataset index:
# SwissRoll = 0
# Peaks = 1
# GMM = 2
index = 2

x = loadmat(filenames[index])
Xtrain, Ytrain,= x['Yt'].T, x['Ct'].T
```

```python
print("Xtrain", Xtrain.shape)
print(Xtrain)
```

```python
print("Ytrain", Ytrain.shape)
print(Ytrain)
```

```python
labels = [np.argmax(y) for y in Ytrain]
print("Labels", np.shape(labels))
print(labels)
```

```python
(unique, counts) = np.unique(labels, return_counts=True)
histogram = dict(zip(unique, counts))
print("Histogram")
print(histogram)

```

```python
sns.jointplot(x=Xtrain.T[0], y=Xtrain.T[1], hue=labels)
```
